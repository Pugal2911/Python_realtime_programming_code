```python
#import random package:
import random

#create a empty list to store a correct word:
lst = []
Wordfile = open("word.txt","r")

for word in Wordfile:
    result = word.rstrip("\n")
    lst.append(result.lower())

#to make a computer to guess a word:
computer = random.choice(lst)
start = False

#run a while loop to start a game:
while start == False:
    game_name = input("Enter a game name to start the game (The game name is SEEKER) : ")
    if game_name.upper() == "SEEKER":
        print("Instruction to play a game : ")
        print(">>> There are 5 letter in a word, the user have to find it!!!")
        print(">>> And also there are 5 turns to find a word to be given!!!")
        start = True
        
    else:
        print('Invalid')

start = False

#then create a empty list to store a misplaced, incorrect word:
misplaced = []
incorrect = []
max_turns = 5
turns_taken = 0

#to run a while loop to guess a word from player:
while turns_taken<=5:
    guess = input("Enter a word with 5 letter").lower()

    if len(guess) != len(computer) or not guess.isalpha():
        continue

    index = 0
    for i in guess:
        if i == computer[index]:
            print(i, end = " ")
            if i in misplaced:
                misplaced.remove(i)

        elif i in computer:
            if i not in misplaced:
                misplaced.append(i)
            print("_", end = " ")

        else:
            if i not in incorrect:
                incorrect.append(i)
            print('_', end = " ")
        index += 1

    print("\n")
    print("Mismatched guess : ", misplaced)
    print("Incorrect : ", incorrect)
    
    

    turns_taken += 1

    if guess == computer:
        print("You won the game")
        break

    elif turns_taken == max_turns:
        print("You can lose the game with exceeds of turns")
        print("The correct guess is", computer)
        break

    print("You had", max_turns - turns_taken, "turns left")
```
